```python
import pandas as pd
import matplotlib.pyplot as plt

df_reg = pd.read_csv("/Users/giovanni-lunetta/stat_4185/final_project/data/all_star_reg.csv")
df_advanced = pd.read_csv("/Users/giovanni-lunetta/stat_4185/final_project/data/all_star_advanced.csv")
df_hall_of_fame = pd.read_csv("/Users/giovanni-lunetta/stat_4185/final_project/data/hall_of_fame.csv")
df_mvp = pd.read_csv("/Users/giovanni-lunetta/stat_4185/final_project/data/mvp.csv")
```

```python
df = pd.merge(df_reg, df_advanced[['Player', 'WS', 'ORtg', 'DRtg', 'VORP', 'PER', 'TRB%', 'AST%', 'TOV%', 'USG%']], on='Player', how='left')
```

```python
df.shape
```

```python
# extract the list of Hall of Fame players
hof_players = set(df_hall_of_fame['Player'])

# create a new column in df indicating whether each player is in the Hall of Fame or not
df['HOF'] = df['Player'].apply(lambda x: 1 if x in hof_players else 0)
```

```python
mvp_players = set(df_mvp['Player'])

df['MVP'] = df ['Player'].apply(lambda x: 1 if x in mvp_players else 0)
```

```python
df.dtypes
```

```python
df = df.drop(['Rk', 'Team', '-9999', 'G', 'GS', 'From', 'To', 'Age'], axis=1)
```

```python
# use boolean indexing to filter the rows where the "WS" column has null values
null_ws_rows = df[df['WS'].isnull()]

# drop the rows where "WS" is null
df = df.dropna(subset=['WS'])

df.shape
```

```python
print((df.isnull().sum() * 100)/ len(df))
```

```python
null_3p_rows = df[df['3P'].isnull()]
null_3p_rows
```

```python
# replace null values in "3P", "3PA" and "3P%" columns with zeros
df['3P'] = df['3P'].fillna(0)
df['3PA'] = df['3PA'].fillna(0)
df['3P%'] = df['3P%'].fillna(0)
```
