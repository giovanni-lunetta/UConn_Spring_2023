```python
import pandas as pd
import matplotlib.pyplot as plt

df_advanced = pd.read_csv("/Users/giovanni-lunetta/stat_4185/final_project/2022-2023_season/data/advanced.csv")
```

```python
print(df_advanced.shape)
```

```python
df_advanced = df_advanced[df_advanced['MP'] >= 12]
df_advanced = df_advanced[df_advanced['G'] >= 21]
```

```python
df_advanced.head()
```

```python
df_advanced = df_advanced.drop(['Rk', 'Team', '-9999', 'GS', 'From', 'To'], axis=1)
```

```python
print((df_advanced.isnull().sum() * 100)/ len(df_advanced))
```

```python
nulrl_ows = df_advanced[df_advanced['FT%'].isnull()]
nulrl_ows
```

```python
# replace null values in "3P", "3PA" and "3P%" columns with zeros
df_advanced['FT%'] = df_advanced['FT%'].fillna(0)
df_advanced['2P%'] = df_advanced['2P%'].fillna(0)
df_advanced['3P%'] = df_advanced['3P%'].fillna(0)
```

```python
print((df_advanced.isnull().sum() * 100)/ len(df_advanced))
```

```python
# Filter the dataframe where 'Pos' is either 'G-F' or 'F-G'
df_gf = df_advanced[df_advanced['Pos'].isin(['G-F'])]

df_gf
```

```python
# Filter the dataframe where 'Pos' is either 'G-F' or 'F-G'
df_fg = df_advanced[df_advanced['Pos'].isin(['F-G'])]

df_fg
```

```python
# Define the mapping dictionary
pos_mapping = {
    'G': 1,
    'G-F': 2,
    'F-G': 3,
    'F': 4,
    'F-C': 5,
    'C-F': 6,
    'C': 7
}

# Replace the values in the 'Pos' column using the mapping dictionary
df_advanced['Pos'] = df_advanced['Pos'].replace(pos_mapping)
```
